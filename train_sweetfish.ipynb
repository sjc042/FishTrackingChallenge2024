```python
# Generate the complete annotated video after detection and tracking

# Define width and height parameters of the video
WIDTH = 2456
HEIGHT = 2058

# Load the YOLOv8 model
model = YOLO('/content/drive/MyDrive/Colab_Notebooks/IPL_Research/Sweetfish_Project/data/best.pt')

# Open the video file
video_path = "/content/drive/MyDrive/Colab_Notebooks/IPL_Research/Sweetfish_Project/data/development.mp4"
cap = cv2.VideoCapture(video_path)


# Instantiate the output video model
output = cv2.VideoWriter("/content/drive/MyDrive/Colab_Notebooks/IPL_Research/Sweetfish_Project/data/track_output_new.avi", cv2.VideoWriter_fourcc(*'MPEG'), 30, (WIDTH, HEIGHT))

counter = 0;
# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
      # Run YOLOv8 tracking on the frame, persisting tracks between frames
      results = model.track(frame, conf=0.695, persist=True, tracker="bytetrack.yaml")

      # Visualize the results on the frame
      annotated_frame = results[0].plot()

      output.write(annotated_frame)
      counter += 1
    else:
      print("finished tracking")
      break

cv2.destroyAllWindows()
output.release()
cap.release()
```

```python
# Generate the tracking results in MOT format and save them in txt file

# Define width and height parameters of the video
WIDTH = 2456
HEIGHT = 2058

# Load the YOLOv8 model
model = YOLO('/content/drive/MyDrive/Colab_Notebooks/Sweetfish_Project/data/best_new.pt')

# Open the video file
video_path = "/content/drive/MyDrive/Colab_Notebooks/Sweetfish_Project/data/development.mp4"
cap = cv2.VideoCapture(video_path)

output_file = os.path.join(ROOT_DIR ,"answer_iou_0.7_new.txt")

frame_num = 1
with open(output_file, 'w') as answer_txt_file:
  # Loop through the video frames
  while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
      # Run YOLOv8 tracking on the frame, persisting tracks between frames
      results = model.track(frame, iou=0.7, persist=True, tracker="bytetrack.yaml")

      boxes = results[0].boxes.xywh.cpu()
      track_ids = results[0].boxes.id.int().cpu().tolist()
      confis = results[0].boxes.conf.cpu()

      # y and x and h and w are swapped because of the format
      # the mot format is for the transposed video
      for box, track_id, confi in zip(boxes, track_ids, confis):
        x, y, w, h = np.array(box)
        y_mot = y - h/2
        x_mot = x - w/2

        answer_txt_file.write(str(frame_num) + " " + str(np.array(track_id)) + " " + str(y_mot) + " " + str(x_mot) + " " + str(h) + " " + str(w) + " "
                                + str(np.array(confi)) + " " + "-1" + " " + "-1" + " " + "-1" + "\n")
      frame_num = frame_num + 1
    else:
      print("finished")
      print(frame_num)
      break

cv2.destroyAllWindows()
cap.release()
```
